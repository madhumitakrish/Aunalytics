```python
#Importing libraries required
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading the train dataset
df=pd.read_csv('au_train.csv',sep=',',na_values='?',skipinitialspace=True)
display(df.head(30))
display(df.describe())
display(df.dtypes)
```

```python
#Total number of missing values
print(df.isnull().sum())

#Percentage of missing values
perofmissing = (df.isnull().sum()/len(df))*100
print('Percentage of missing values \n',perofmissing)

#Dropping Nan values

df=df.dropna()
print(len(df))
```

```python
# Total number of Employees
print('Total number of Employees:',len(df))

# Number of employees with income less than or equal to 50K
payless= np.sum(df['class']=='<=50K')
print('Number of employees with income less than or equal to 50K',payless)
# Number of employees with income greater than 50K
paymore= np.sum(df['class']=='>50K')
print('Number of employees with income greater than 50K',paymore)

#Data is imbalanced
```

```python
# Plotting the income by education
plt.figure(figsize=(5,5))
ax = sns.countplot(x = 'education', hue = 'class', data = df, saturation=1)
ax.set_title('Income by Education')
ax.set_xlabel('Education')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
ax.set_ylabel('Number of Employees')
ax.legend(loc=1, title='Income')

plt.tight_layout()
plt.show()
```

The above Income by Education graph displays a few noteable occurances. Most number of employees can be seen to be high school graduates or have attended some college. Most number of employees with higher pay have a Bachelors closely followed by high school grads.

```python

# Plotting the income by Occupation
plt.figure(figsize=(5,5))
ax = sns.countplot(x = 'occupation', hue = 'class', data = df, saturation=1)
ax.set_title('Income by Occupation')
ax.set_xlabel("Occupation")
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
ax.set_ylabel("Number of Employees")
ax.legend(loc=1, title='Income')
plt.tight_layout()
plt.show()
```

The above graph shows least number of people employed in Armed Forces and Private-house-service jobs. A large number of other occupations seem to have distributed number of employees. In case of clerical jobs, craft-repair and other services the total number of employees who earn less than 50k are very high.

```python
# Plotting the income by Marital Status
plt.figure(figsize=(5,5))
ax = sns.countplot(x = 'marital-status', hue = 'class', data = df, saturation=1)
ax.set_title('Income by Marital Status')
ax.set_xlabel('Marital-Status')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
ax.set_ylabel('Number of Employees')
ax.legend(loc=1, title='Income')

plt.tight_layout()
plt.show()
```

The graph shows most number of Employees to be never married or married with spouse. More number of employees who are married with spouse are seen to have higher income.

```python
# Plotting the income by race
plt.figure(figsize=(5,5))
ax = sns.countplot(x = 'race', hue = 'class', data = df, saturation=1)
ax.set_title('Income by Race ')
ax.set_xlabel('Race')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
ax.set_ylabel('Number of Employees')
ax.legend(loc=1, title='Income')

plt.tight_layout()
plt.show()
```

The data consists of employees of the white race predominantly and a large disparity can be seen in the number of people in different classes of income.

```python
# Plotting the income by Age
plt.figure(figsize=(12,7))
agedf=df.sort_values(by=['age'])

ax = sns.countplot(x = 'age', hue = 'class', data = agedf, saturation=1)
ax.set_title('Income by Age ')
ax.set_xlabel('Age')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
ax.set_ylabel('Number of Employees')
ax.legend(loc=1, title='Income')

plt.tight_layout()
plt.show()
```

The above graph is sorted by age and the number of Employees is skewed to the right which shows that employee count decreases with age. It can also be seen that people whose ages are between 30 to 50 have higher income when compared to others. Age can be considered one of the important factors for modelling

```python
# Plotting the income by sex
plt.figure(figsize=(5,5))

ax = sns.countplot(x= 'sex', hue= 'class', data= df, saturation= 1)
ax.set_title('Income by Sex ')
ax.set_xlabel('Sex')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha= "right")
ax.set_ylabel('Number of Employees')
ax.legend(loc= 1, title='Income')

plt.tight_layout()
plt.show()
```

```python
sex_df=df[df['sex']=='Female']
print(sex_df.shape)
total_df=sex_df[sex_df['class']=='>50K']
print(total_df.shape)
```

The above graph shows income and number of Employees by sex. It can be seen that the number of male employees is much more than the number of female employees. One other major difference is the number of male employees earining more that 50k as compared to number of female employees earning more that 50k. By calculating only 11.36% of the total number of women employees earn greater than 50k.

```python
# Plotting the income by workclass
plt.figure(figsize=(5,5))

ax = sns.countplot(x= 'workclass', hue= 'class', data= df, saturation= 1)
ax.set_title('Income by Work Class ')
ax.set_xlabel('Work Class')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha= "right")
ax.set_ylabel('Number of Employees')
ax.legend(loc= 1, title='Income')

plt.tight_layout()
plt.show()
```
